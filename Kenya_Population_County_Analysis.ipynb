# Load Data

```python
data = pd.read_csv("Datasets/Population-households-density-by-county.csv", index_col=False)
```

```python
#Remove trailing spaces
data.columns = data.columns.str.strip()
```

```python
data
```

# Bar graph of Counties against Population

```python
data['Total'] = data['Total'].str.replace(',', '').astype(int)
counties = data['County']
```

```python
plt.bar(counties, data['Total'])
# Adding labels and title
plt.xlabel('County')
plt.ylabel('Total Population')
plt.title('Total Population by County')
# Rotating the x-axis labels for better readability
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```


# Bar graph of counties againt Total Households
